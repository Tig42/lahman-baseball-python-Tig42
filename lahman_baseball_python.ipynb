```python
query ="""
select c.playerid,s.schoolname,p.namefirst,p.namelast,sum(sr.salary)as total_salary
from schools as s right join collegeplaying as c
    on s.schoolid=c.schoolid right join people as p 
    on c.playerid=p.playerid right join salaries as sr
    on p.playerid=sr.playerid
where s.schoolname ilike '%%vanderbilt%%'
group by  c.playerid,s.schoolname,p.namefirst,p.namelast
order by total_salary 
"""
salary=pd.read_sql(query,engine)
salary
```

3. a. From 1970 â€“ 2016, what is the largest number of wins for a team that did not win the world series?  

```python

query ="""
select teamid,a.yearid,wswin, max(w) as max_wins
from awardsshareplayers a left join teams as t
   on a.yearid=t.yearid
where a.yearid between 1970 and 2016 and wswin='N'
Group by teamid,a.yearid,wswin
order by max_wins DESC
"""
win_team=pd.read_sql(query,engine)
win_team
```

4. Which managers have won the TSN Manager of the Year award in both the National League (NL) and the American League (AL)? Give their full name and the teams that they were managing when they won the award.

```python

query ="""
SELECT p.namefirst, p.namelast, m.teamid, a.playerid, a.lgid
FROM awardsmanagers AS a
INNER JOIN people AS p
ON a.playerid = p.playerid
INNER JOIN managers AS m
ON m.playerid = a.playerid
WHERE awardid = 'TSN Manager of the Year'
AND a.lgid = 'AL' OR a.lgid = 'NL'
GROUP BY p.namefirst, p.namelast, a.playerid, a.lgid, m.teamid
ORDER BY a.playerid;
"""
TSN_manager=pd.read_sql(query,engine)
TSN_manager.head()

```
